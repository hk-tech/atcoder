```python
import math

def combinations_count(n, r):
    if n <= 1:
        return 0
    return math.factorial(n) // (math.factorial(n - r) * math.factorial(r))
```

```python
def combinations_mod(n, r, mod=1000000007):
    """Returns nCr in mod."""
    r = min(r, n - r)
    combs = 1
    for i, j in zip(range(n - r + 1, n + 1), range(1, r + 1)):
        combs *= (i % mod) * pow(j, mod - 2, mod)
        combs %= mod
    return combs
```

## 二分木

```python
class BinaryTree:
    def __init__(self, root = None):
        self.value  = root
        self.left  = None
        self.right = None

    def insert_left(self, insert):
        if self.left == None:
            self.left = BinaryTree(insert)
        else:
            t = BinaryTree(insert)
            t.left = self.left
            self.left = t
    
    def insert_right(self, insert):
        if self.right == None:
            self.right = BinaryTree(insert)
        else:
            t = BinaryTree(insert)
            t.right = self.right
            self.right = t
    
    def get_left(self):
        return self.left
    
    def get_right(self):
        return self.right
    
    def size(self):
        count = 1
        if self.left:
            count += self.left.size()
        if self.right:
            count += self.right.size()
        return count
```

## UnionFind

```python
N, M, K=map(int,input().split())
AB=[list(map(int,input().split()))for _ in range(M)]
CD=[list(map(int,input().split()))for _ in range(K)]
import sys;sys.setrecursionlimit(10**9)
class UnionFind():
    def __init__(self, n):
        self.n = n
        self.parents = [-1] * n
 
    def find(self, x):
        if self.parents[x] < 0:
            return x
        else:
            self.parents[x] = self.find(self.parents[x])
            return self.parents[x]
 
    def union(self, x, y):
        x = self.find(x)
        y = self.find(y)
 
        if x == y:
            return
 
        if self.parents[x] > self.parents[y]:
            x, y = y, x
 
        self.parents[x] += self.parents[y]
        self.parents[y] = x
 
    def size(self, x):
        return -self.parents[self.find(x)]
 
    def same(self, x, y):
        return self.find(x) == self.find(y)
 
    def members(self, x):
        root = self.find(x)
        return [i for i in range(self.n) if self.find(i) == root]
 
    def roots(self):
        return [i for i, x in enumerate(self.parents) if x < 0]
 
    def group_count(self):
        return len(self.roots())
 
    def all_group_members(self):
        return {r: self.members(r) for r in self.roots()}
 
    def __str__(self):
        return '\n'.join('{}: {}'.format(r, self.members(r)) for r in self.roots())

u=UnionFind(N)
for a,b in AB:
    u.union(a-1,b-1)
    
ans=[u.size(i)-1 for i in range(N)]
 
for a,b in AB:
  if u.same(a-1,b-1):
    ans[a-1]-=1
    ans[b-1]-=1
for c,d in CD:
  if u.same(c-1,d-1):
    ans[c-1]-=1
    ans[d-1]-=1
print(*ans)
```

## 2点を通る直線の方程式

```python
def makeLinearEquation(x1, y1, x2, y2):
	line = {}
	if y1 == y2:
		# y軸に平行な直線
		line["y"] = y1
	elif x1 == x2:
		# x軸に平行な直線
		line["x"] = x1
	else:
		# y = mx + n
		line["m"] = (y1 - y2) / (x1 - x2)
		line["n"] = y1 - (line["m"] * x1)
	return line

eq = makeLinearEquation(1, 1, 7, -2)
eq
```

```python
N = int(input())
dp = [0] * (N+1) #見ているところまでで結果が1になるパターン数を保管。
dp[0] = 1 
for i in range(N):
    s = input()
    if s == 'OR':
        dp[i+1] = 2 ** (i+1) + dp[i] #最新がORであれば、1の時それまではなんでもいいので、2^(i+1)を足す。0のときは1じゃないといけないためdp[i]を足す
    else:
        dp[i+1] = dp[i] #最新がANDならば、それまでは1じゃないといけないためdp[i]を足す。全部ANDの場合は初期の1がずっと引き継がれる

print(dp[N])
```

## N = 10^4 で O(N^2)の計算量を通せる（PyPy)
https://atcoder.jp/contests/abc189/tasks/abc189_c
- 範囲を絞って全探索する必要あり

```python
N = int(input())
A = list(map(int, input().split()))
ans = 0
for i in range(N):
    m = 10 ** 9 
    for j in range(i, N):
        m = min(m, A[j])
        ans = max(ans, m * (j-i+1))
 
print(ans)
```

### 高速化（PyPy）

https://atcoder.jp/contests/abc179/tasks/abc179_c

```python
N = int(input())

lis = [0] * N
ans = 0

for i in range(1, N + 1):
    for j in range(1, N//i+1):
        if i * j <= N - 1:
            lis[i*j] += 1

print(sum(lis))
```

## 整数問題

### Nだけ与えられる系の問題では約数を疑おう
- https://atcoder.jp/contests/abc190/tasks/abc190_d

整数からなる公差 1 の等差数列のうち、総和が N であるものはいくつあるでしょうか？

等差和の公式で [A, A+1, ..., B-1, B]がNとなるケースを考えれば良い。

(A+B) * (B - A + 1)/2 = N なので

(A+B) * (B - A + 1) = 2N

2Nの約数を求める。偶奇が異なるペアしか成り立たないので、約数のうち奇数の数を数えるとそれが答え

```python
N = int(input())

def make_divisors(n):
    divisors = []
    for i in range(1, int(n**0.5)+1):
        if n % i == 0:
            divisors.append(i)
            if i != n // i:
                divisors.append(n//i)
 
    divisors.sort()
    return divisors

d = make_divisors(2 * N)

ans = 0
for x in d:
    if x % 2 == 1:
        ans += 1

print(ans*2)
```
